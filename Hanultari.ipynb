```python
page = 1
url = f'https://www.mcfamily.or.kr/web/board/program.php?page={page}&Scentercd=&Sstatus=&Sword='
```

```python
title_list = []

for page in range(1, 478):
  t_url = url.format(page)
  response = requests.get(t_url)
  soup = bs(response.content, 'html.parser')
  titles = soup.select('#sub_container > div.borad_wrap > div.apply_list_wrap > dl > dt > a')

  for title in titles:
    title_list.append(title.get_text())
```

```python
title_list
```

```python
df = pd.DataFrame(title_list)
df
```

```python
gu_list = []

for page in range(1, 478):
  t_url = url.format(page)
  response = requests.get(t_url)
  soup = bs(response.content, 'html.parser')
  gus = soup.select('#sub_container > div.borad_wrap > div.apply_list_wrap > dl.list_sect > dt > ul > li:nth-child(1) > span')

  for gu in gus:
    gu_list.append(gu.get_text())
```

```python
gu_list
```

```python
upload_date_list = []

for page in range(1, 478):
  t_url = url.format(page)
  response = requests.get(t_url)
  soup = bs(response.content, 'html.parser')
  upload_date = soup.select('#sub_container > div.borad_wrap > div.apply_list_wrap > dl > dt > ul > li:nth-child(2) > span')

  for ud in upload_date:
    upload_date_list.append(ud.get_text())
```

```python
upload_date_list
```

```python
#sub_container > div.borad_wrap > div.apply_list_wrap > dl > dd > p:nth-child(1)
```

```python
accept_date_list = [] # 접수 기간
prog_date_list = [] # 진행 기간

for page in range(1, 478):
  t_url = url.format(page)
  response = requests.get(t_url)
  soup = bs(response.content, 'html.parser')
  accept_date = soup.select('#sub_container > div.borad_wrap > div.apply_list_wrap > dl > dd > p:nth-child(1)')
  prog_date = soup.select('#sub_container > div.borad_wrap > div.apply_list_wrap > dl > dd > p:nth-child(2)')

  for ad in accept_date:
    accept_date_list.append(ad.get_text())
  for prd in prog_date:
    prog_date_list.append(prd.get_text())
```

```python
data = {
  'Title': title_list,
  'Districts': gu_list,
  'Upload_date': upload_date_list,
  'Accept_date': accept_date_list,
  'Progress_date': prog_date_list
  
}

df = pd.DataFrame(data)
df
```

```python
file_name = 'mcfamily_program.csv'
df.to_csv(file_name, index=False)
```
